```python

# Calculate average scores for each model
average_scores = {model: sum(scores) / len(scores) for model, scores in models_scores.items()}

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(average_scores.keys(), average_scores.values(), color=['blue', 'green', 'orange', 'red'])

# Adding labels and title
plt.ylabel('Average BERTScore')
plt.title('Average BERTScore by Model')
plt.ylim(0.5, 0.8)  # Adjust y-axis limits for better visualization
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python

# Calculate average score for each book across all models
scores_per_book = list(zip(*models_scores.values()))
average_scores_per_book = [sum(scores) / len(scores) for scores in scores_per_book]

# Plotting
plt.figure(figsize=(14, 6))
plt.bar(books, average_scores_per_book, color='skyblue')

# Adding labels and title
plt.ylabel('Average BERTScore')
plt.title('Average BERTScore per Book Across Models')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Show plot
plt.show()
```
