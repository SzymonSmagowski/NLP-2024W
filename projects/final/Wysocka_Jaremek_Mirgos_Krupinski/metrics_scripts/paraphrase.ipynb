```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for file in file_names:
    print(file)
    df_model_answers = pd.read_table(path_to_model_answers + '/' + file + '.txt', sep='\t', header=None)

    result = pd.DataFrame(columns=['answer'])

    for index, r in df_model_answers.iterrows():
        print(file, index, r[0])
        result = result.append({'answer': 
                                client.generation("Paraphrase following sentence five times, differently:" + 
                                                    r[0], max_length=1000)['generated_text'] + '\n==='}, ignore_index=True)

    print(result)
    result.to_csv(path_to_model_answers + '/' + file + '_paraphrased.txt', sep='\t', header=None, index=None)
```

```python
def clean_up(df):
    df.iloc[:,0] = df.iloc[:,0].str.replace('Paraphrase following sentence five times, differently:', '')
    df.iloc[:,0] = df.iloc[:,0].str.replace(' 1. ', '\n')
    df.iloc[:,0] = df.iloc[:,0].str.replace('2. ', '')
    df.iloc[:,0] = df.iloc[:,0].str.replace('3. ', '')
    df.iloc[:,0] = df.iloc[:,0].str.replace('4. ', '')
    df.iloc[:,0] = df.iloc[:,0].str.replace('5. ', '')

    return df
```

```python
for file in file_names:
    print(file)
    df = pd.read_table(path_to_model_answers + '/' + file + '_paraphrased.txt', sep='\t', header=None)
    df = clean_up(df)
    df.to_csv(path_to_model_answers + '/' + file + '_paraphrased.txt', sep='\t', header=None, index=None)
```
